```python
# Combine Data by Market Cap

df_crypto = pd.DataFrame({"Date": df_btc["Date"],
                          "BTC Cap":df_btc_2y["Market Cap"], 
                          "ETH Cap": df_eth_2y["Market Cap"],
                          "XRP Cap": df_xrp_2y['Market Cap'],
                          "BCH Cap": df_bch_2y['Market Cap'],
                          "EOS Cap": df_eos_2y['Market Cap'],
                          "XLM Cap": df_xlm_2y['Market Cap'],
                          "LTC Cap": df_ltc_2y['Market Cap'],
                          "ADA Cap": df_ada_2y['Market Cap'],
                          "XMR Cap": df_xmr_2y['Market Cap'],
                          "TRX Cap": df_trx_2y['Market Cap'],
                         })
# Combine data by Volume
df_crypto_volume = pd.DataFrame({"Date": df_btc_2y["Date"],
                          "BTC Vol":df_btc_2y["Volume"], 
                          "ETH Vol": df_eth_2y["Volume"],
                          "XRP Vol": df_xrp_2y['Volume'],
                          "BCH Vol": df_bch_2y['Volume'],
                          "EOS Vol": df_eos_2y['Volume'],
                          "XLM Vol": df_xlm_2y['Volume'],
                          "LTC Vol": df_ltc_2y['Volume'],
                          "ADA Vol": df_ada_2y['Volume'],
                          "XMR Vol": df_xmr_2y['Volume'],
                          "TRX Vol": df_trx_2y['Volume'],
                         })

df_crypto_volume.head()
```

```python
# Arrange Cryptos by market cap
most_update = df_crypto.loc[0]
most_update_drop = most_update.drop("Date")
most_update_drop

most_update_sort = most_update_drop.sort_values(ascending=False )

most_update_sort

top_five = most_update_sort.head()

top_five_names = [top_five.index[i] for i in range(0,5)]
top_five_values = [top_five.values[i] for i in range(0,5)]

top_five_names
```

```python
# Graph Data

plt.rcdefaults()
fig, ax = plt.subplots()

# Example data
y_pos = np.arange(len(top_five_names))
error = np.random.rand(len(top_five_names))

ax.barh(y_pos, top_five_values, xerr=error, align='center',
        color='blue', ecolor='black')
ax.set_yticks(y_pos)
ax.set_yticklabels(top_five_names)
ax.invert_yaxis()  # labels read top-to-bottom
ax.set_xlabel('Crypto Market Cap')
ax.set_title('Top 5 Crypto for today')

plt.show()
```

```python
#Percentage change and create a new data frame 

#Bitcoin dataframe for 2 years data
df_btc_2y["Daily Change Percent(%)"] = ((df_btc_2y["Close"] - df_btc_2y["Close"].shift(-1))/df_btc_2y["Close"].shift(-1))*100
df_btc_2y = df_btc_2y.dropna()

#df_btc_1y["Daily Change Percent(%)"] = ((df_btc_1y["Close"] - df_btc_1y["Close"].shift(-1))/df_btc_1y["Close"].shift(-1))*100
#df_btc_6m["Daily Change Percent(%)"] = ((df_btc_6m["Close"] - df_btc_6m["Close"].shift(-1))/df_btc_6m["Close"].shift(-1))*100
#df_btc_1m["Daily Change Percent(%)"] = ((df_btc_1m["Close"] - df_btc_1m["Close"].shift(-1))/df_btc_6m["Close"].shift(-1))*100


#Etheurum

df_eth_2y["Daily Change Percent(%)"] = ((df_eth_2y["Close"] - df_eth_2y["Close"].shift(-1))/df_eth_2y["Close"].shift(-1))*100

#Ripple

df_xrp_2y["Daily Change Percent(%)"] = ((df_xrp_2y["Close"] - df_xrp_2y["Close"].shift(-1))/df_xrp_2y["Close"].shift(-1))*100

# Bitcoin Cash

df_bch_2y["Daily Change Percent(%)"] = ((df_bch_2y["Close"] - df_bch_2y["Close"].shift(-1))/df_bch_2y["Close"].shift(-1))*100

# EOS

df_eos_2y["Daily Change Percent(%)"] = ((df_eos_2y["Close"] - df_eos_2y["Close"].shift(-1))/df_eos_2y["Close"].shift(-1))*100

df_btc_2y

```

```python
# resample daily change percentage weekly
Index = pd.DatetimeIndex(start='2016-11-03', end='2018-11-09', freq="D")
close = df_btc_2y["Close"].tolist()
s_btc = pd.DataFrame({"Close": close}, index=Index)

#s_btc.index = pd.to_datetime(s_btc.index,format="%Y-%m-%d")

s_btc = s_btc.resample(rule="M",how=np.std)
s_btc = s_btc.rename(columns={"Close":"Monthly Deviation"})

%matplotlib notebook
s_btc_dev = s_btc["Monthly Deviation"]
s_btc_dev.plot(kind='hist', title= "Bitcoin Deviation Distribution Based on Closing Price", color="red", alpha=0.5)

plt.savefig("graphs/deviation/bitcoin_dev_dist.png")
```
