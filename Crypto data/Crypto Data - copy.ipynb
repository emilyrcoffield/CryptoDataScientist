```python
# scraping data
scraper_btc = CmcScraper("btc")
scraper_eth = CmcScraper("eth")
scraper_xrp = CmcScraper("xrp")
scraper_eos = CmcScraper("eos")
scraper_bch = CmcScraper("bch")
scraper_xlm = CmcScraper("xlm")
scraper_ltc = CmcScraper("ltc")
scraper_ada = CmcScraper("ada")
scraper_xmr = CmcScraper("xmr")
scraper_trx = CmcScraper("trx")


# converting to csv

scraper_btc.export_csv("btc_all_time.csv")
scraper_eth.export_csv("eth_all_time.csv")
scraper_xrp.export_csv("xrp_all_time.csv")
scraper_eos.export_csv("eos_all_time.csv")
scraper_bch.export_csv("bch_all_time.csv")
scraper_xlm.export_csv("xlm_all_time.csv")
scraper_ltc.export_csv("ltc_all_time.csv")
scraper_ada.export_csv("ada_all_time.csv")
scraper_xmr.export_csv("xmr_all_time.csv")
scraper_xmr.export_csv("trx_all_time.csv")
```
